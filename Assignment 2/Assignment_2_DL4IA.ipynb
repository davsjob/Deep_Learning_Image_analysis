```python
import numpy as np
import matplotlib.pyplot as plt
from load_mnist import load_mnist
from A2_DL4IA import *

```

## Importing and splitting the MNIST dataset

```python
x_train, y_train, x_test, y_test  = load_mnist()

```

## 1-Layer Network

```python
one_layer = {
    "layer1": {"nodes": 256, "activation": "relu"},
    "output": {"nodes": 10, "activation": "softmax"}
}
network_1l = FeedForwardBackPropagationNetwork(one_layer)
network_1l.train_model(x_train, y_train, x_test, y_test, mu=0, sigma=0.01, iterations=100, learning_rate=0.01, batch_size=64)
```

```python
network_1l.training_curve_plot("One Layer Neural Network", network_1l.costs_training, network_1l.costs_validation, network_1l.accuracies_training, network_1l.accuracies_validation )
```

```python
network_1l.plot_weights()
```

## 2-layer network

```python
model = {
        "layer1": {"nodes": 2*256, "activation": "relu"},
        "layer2": {"nodes": 2*128, "activation": "relu"},
        "output": {"nodes": 10, "activation": "softmax"}
    }
network = FeedForwardBackPropagationNetwork(model)
network.train_model(x_train, y_train, x_test, y_test, mu=0, sigma=0.01, iterations=100, learning_rate=0.01, batch_size=64)
```

```python
network.training_curve_plot("Training plot for 2 layer network", network.costs_training, network.costs_validation, network.accuracies_training, network.accuracies_validation)
```

```python
network.plot_weights()
```
